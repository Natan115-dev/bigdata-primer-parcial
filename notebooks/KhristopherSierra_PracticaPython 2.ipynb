# Ejercicio 1: Clasificación de edad  

**Enunciado:** determinar si una persona es niño, adolescente, adulto o adulto mayor según su edad.  

**Lógica:** se evalúa la edad con condiciones if, elif y else. El primer rango que se cumple determina la categoría.

```python
def clasificar_edad(edad):
    if edad < 13:
        return "Niño"
    elif edad <= 17:
        return "Adolescente"
    elif edad <= 59:
        return "Adulto"
    else:
        return "Adulto mayor"
edad = 21
categoria = clasificar_edad(edad)
print(f"Edad: {edad} años")
print(f"Categoría: {categoria}")
```

# Ejercicio 2: Conversor de temperatura  

**Enunciado:** convertir una temperatura de grados Celsius a Fahrenheit.  

**Lógica:** se utiliza la fórmula F = (C × 9/5) + 32 para convertir la temperatura.  

```python
def convertir_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit
celsius = 30
resultado = convertir_fahrenheit(celsius)
print(f"Temperatura en Celsius: {celsius}°C")
print(f"Temperatura en Fahrenheit: {resultado}°F")
```

# Ejercicio 3: Calculadora básica  

**Enunciado:** realizar operaciones matemáticas básicas entre dos números.  

**Lógica:** se utilizan condiciones para identificar la operación seleccionada y validar que no se divida entre cero.

```python
def calculadora(num1, num2, operacion):
    if operacion == "+":
        return num1 + num2
    elif operacion == "-":
        return num1 - num2
    elif operacion == "*":
        return num1 * num2
    elif operacion == "/":
        if num2 == 0:
            return "Error: no se puede dividir entre cero"
        return num1 / num2
    else:
        return "Operación no válida"
num1 = 20
num2 = 5
operacion = "/"
resultado = calculadora(num1, num2, operacion)
print(f"Número 1: {num1}")
print(f"Número 2: {num2}")
print(f"Operación: {operacion}")
print(f"Resultado: {resultado}")
```

# Ejercicio 4: Número par o impar

**Enunciado:** determinar si un número entero es par o impar.

**Lógica:** se utiliza el operador módulo (%) para obtener el residuo de la división entre 2. Si el residuo es igual a 0, el número es par; de lo contrario, es impar.

```python
def verificar_numero(numero):
    if numero % 2 == 0:
        return "Par"
    else:
        return "Impar"

numero = 15
resultado = verificar_numero(numero)

print(f"Número: {numero}")
print(f"Resultado: {resultado}")
```

# Ejercicio 7: Promedio de calificaciones

**Enunciado:** calcular el promedio de 5 calificaciones e indicar si el estudiante aprobó o reprobó.

**Lógica:** se suman las 5 calificaciones, se calcula el promedio y se evalúa con una condición `if` si el promedio es mayor o igual a 70.

```python
def calcular_promedio(calificaciones):
    suma = sum(calificaciones)
    promedio = suma / len(calificaciones)

    if promedio >= 70:
        resultado = "Aprobó"
    else:
        resultado = "Reprobó"

    return promedio, resultado


calificaciones = [85, 70, 90, 60, 75]

promedio, resultado = calcular_promedio(calificaciones)

print(f"Calificaciones: {calificaciones}")
print(f"Promedio: {promedio}")
print(f"Resultado: {resultado}")
```

# Ejercicio 8: Lista de números positivos y negativos

**Enunciado:** contar cuántos números son positivos, negativos y ceros dentro de una lista de 10 números.

**Lógica:** se recorre la lista usando un ciclo `for` y se evalúa cada número con condiciones `if`, `elif` y `else`.

```python
def contar_numeros(lista):
    positivos = 0
    negativos = 0
    ceros = 0

    for numero in lista:
        if numero > 0:
            positivos += 1
        elif numero < 0:
            negativos += 1
        else:
            ceros += 1

    return positivos, negativos, ceros


numeros = [5, -3, 0, 8, -1, 0, 12, -7, 4, -2]

positivos, negativos, ceros = contar_numeros(numeros)

print(f"Números: {numeros}")
print(f"Positivos: {positivos}")
print(f"Negativos: {negativos}")
print(f"Ceros: {ceros}")
```

# Ejercicio 9: Mayor y menor de una lista

**Enunciado:** encontrar el número mayor y menor de una lista y calcular la diferencia entre ambos.

**Lógica:** se utilizan las funciones `max()` y `min()` para encontrar el valor mayor y menor de la lista. Luego se calcula la diferencia restando ambos valores.

```python
def analizar_lista(lista):
    mayor = max(lista)
    menor = min(lista)
    diferencia = mayor - menor

    return mayor, menor, diferencia


numeros = [15, 8, 23, 4, 42, 16, 9]

mayor, menor, diferencia = analizar_lista(numeros)

print(f"Lista: {numeros}")
print(f"Número mayor: {mayor}")
print(f"Número menor: {menor}")
print(f"Diferencia: {diferencia}")
```

## Explicación de `max()` y `min()`

- `max()` devuelve el número más grande de una lista.
- `min()` devuelve el número más pequeño de una lista.

# Ejercicio 12: Validación de contraseña

**Enunciado:** verificar si una contraseña es válida.

**Lógica:** se evalúan condiciones como longitud mínima, presencia de mayúsculas, minúsculas y números usando funciones como `any()`.

```python
password = "Abc1234567"

valida = (
    len(password) >= 8 and
    any(c.isupper() for c in password) and
    any(c.islower() for c in password) and
    any(c.isdigit() for c in password)
)

print("Contraseña valida" if valida else "Contraseña invalida")
```

# Ejercicio 14: Registro de ventas

**Enunciado:** crear una lista de ventas que contenga cliente, producto, cantidad y precio unitario. Calcular el total por venta, el total general vendido e identificar el cliente con la compra más alta.

**Lógica:** se utiliza una lista de diccionarios para almacenar la información de cada venta. Se recorre la lista con un ciclo para calcular el total de cada venta y acumular el total general. También se compara cada venta para encontrar la compra con el valor más alto.

```python
ventas = [
    {"cliente": "Juan", "producto": "Laptop", "cantidad": 1, "precio_unitario": 15000},
    {"cliente": "Maria", "producto": "Mouse", "cantidad": 2, "precio_unitario": 300},
    {"cliente": "Carlos", "producto": "Monitor", "cantidad": 1, "precio_unitario": 4500}
]

total_general = 0
compra_mas_alta = 0
cliente_mayor_compra = ""

print("Registro de ventas")

for venta in ventas:
    total_venta = venta["cantidad"] * venta["precio_unitario"]
    total_general += total_venta

    print(f"Cliente: {venta['cliente']}")
    print(f"Producto: {venta['producto']}")
    print(f"Total venta: L. {total_venta}")
    print()

    if total_venta > compra_mas_alta:
        compra_mas_alta = total_venta
        cliente_mayor_compra = venta["cliente"]

print(f"Total general vendido: L. {total_general}")
print(f"Cliente con la compra más alta: {cliente_mayor_compra}")
print(f"Monto de la compra más alta: L. {compra_mas_alta}")
```

# Ejercicio 15: Limpieza básica de texto

**Enunciado:** limpiar una frase eliminando espacios al inicio y al final, convertirla a minúsculas, reemplazar espacios dobles por uno solo y contar cuántas palabras contiene.

**Lógica:** se utilizan métodos de cadenas para limpiar y transformar el texto. `strip()` elimina espacios innecesarios al inicio y final, `lower()` convierte todo a minúsculas, `replace()` sustituye espacios dobles por uno simple y `split()` divide la frase en palabras para poder contarlas.

```python
frase = "   Python  es un lenguaje MUY util para Big Data   "

texto_limpio = frase.strip()
texto_limpio = texto_limpio.lower()
texto_limpio = texto_limpio.replace("  ", " ")

palabras = texto_limpio.split()

print(f"Texto original: '{frase}'")
print(f"Texto limpio: '{texto_limpio}'")
print(f"Cantidad de palabras: {len(palabras)}")
```

## Explicación de los métodos utilizados

- **strip()**: elimina los espacios en blanco que se encuentran al inicio y al final de una cadena de texto.

- **lower()**: convierte todos los caracteres del texto a letras minúsculas.

- **replace()**: permite reemplazar una cadena de caracteres por otra dentro del texto. En este ejercicio se utiliza para sustituir espacios dobles por un solo espacio.

- **split()**: divide una cadena en una lista de palabras utilizando los espacios como separador. Esto permite contar cuántas palabras contiene el texto.

# Ejercicio 17: Análisis de ventas con pandas

**Enunciado:** estructurar un DataFrame con 15 ventas para calcular montos totales, agrupar por vendedor y categoría, identificar el producto líder y ordenar los registros.

**Lógica:** se crea un DataFrame con pandas. Se calcula TotalVenta multiplicando cantidad por precio. Se usan groupby(), sum() e idxmax() para obtener los totales y líderes, y sort_values() para el ordenamiento descendente.

```python
import pandas as pd

datos_ventas = {
    "fecha": ["2026-05-01", "2026-05-02", "2026-05-02", "2026-05-03", "2026-05-04",
              "2026-05-04", "2026-05-05", "2026-05-06", "2026-05-06", "2026-05-07",
              "2026-05-08", "2026-05-08", "2026-05-09", "2026-05-10", "2026-05-10"],
    "vendedor": ["Carlos", "Ana", "Luis", "Carlos", "Ana", 
                 "Luis", "Carlos", "Ana", "Luis", "Carlos",
                 "Ana", "Luis", "Carlos", "Ana", "Luis"],
    "producto": ["Laptop", "Mouse", "Teclado", "Monitor", "Mouse",
                 "Teclado", "Laptop", "Monitor", "Laptop", "Mouse",
                 "Teclado", "Monitor", "Laptop", "Mouse", "Teclado"],
    "categoria": ["Tecnología", "Accesorios", "Accesorios", "Tecnología", "Accesorios",
                  "Accesorios", "Tecnología", "Tecnología", "Tecnología", "Accesorios",
                  "Accesorios", "Tecnología", "Tecnología", "Accesorios", "Accesorios"],
    "cantidad": [2, 5, 3, 1, 4, 2, 1, 2, 1, 10, 5, 1, 1, 2, 4],
    "precio_unitario": [800.00, 25.00, 75.00, 180.00, 25.00,
                       75.00, 800.00, 180.00, 800.00, 25.00,
                       75.00, 180.00, 800.00, 25.00, 75.00]
}

df_ventas = pd.DataFrame(datos_ventas)

df_ventas["TotalVenta"] = df_ventas["cantidad"] * df_ventas["precio_unitario"]

print("TOTAL VENDIDO POR VENDEDOR")
total_vendedor = df_ventas.groupby("vendedor")["TotalVenta"].sum().reset_index()
print(total_vendedor)

print("\nPRODUCTO MÁS VENDIDO")
producto_mas_vendido = df_ventas.groupby("producto")["cantidad"].sum().idxmax()
print(f"El producto más vendido es: {producto_mas_vendido}")

print("\nCATEGORÍA CON MAYOR INGRESO")
categoria_mayor_ingreso = df_ventas.groupby("categoria")["TotalVenta"].sum().idxmax()
print(f"La categoría con mayor ingreso es: {categoria_mayor_ingreso}")

print("\nVENTAS ORDENADAS DE MAYOR A MENOR")
df_ordenado = df_ventas.sort_values(by="TotalVenta", ascending=False)
print(df_ordenado)
```

# Ejercicio 18: Limpieza de datos con pandas

**Enunciado:** estructurar un DataFrame con datos incompletos de clientes para contar valores nulos, reemplazar los vacíos de la ciudad por un texto genérico y eliminar registros sin correo.

**Lógica:** se crea un DataFrame con valores nulos (NaN/None). Se detectan y suman los nulos con isnull().sum(), se rellenan las ciudades vacías usando fillna(), y se descartan las filas sin correo mediante dropna(subset=[...]).

```python
import pandas as pd
import numpy as np

datos_clientes = {
    "nombre": ["Carlos Mendieta", "Ana Cruz", "Luis Valle", "Marta Gómez", "Jorge Paz",
               "Elena Solís", "Diego Reyes", "Laura Núñez", "Sofia Meza", "Ramiro Lobo"],
    "edad": [25, np.nan, 30, 22, np.nan, 28, 35, np.nan, 19, 40],
    "correo": ["carlos@mail.com", "ana@mail.com", None, "marta@mail.com", "jorge@mail.com",
               None, "diego@mail.com", "", "none", "ramiro@mail.com"],
    "ciudad": ["Tegucigalpa", "San Pedro Sula", "La Ceiba", np.nan, "Comayagua",
               "Choluteca", np.nan, "Santa Rosa", "Danlí", np.nan],
    "telefono": ["9988-7766", "3322-1100", "8877-6655", None, "7766-5544",
                 None, "9900-1122", "3344-5566", None, "8899-0011"]
}
df_clientes = pd.DataFrame(datos_clientes)

df_clientes["correo"] = df_clientes["correo"].replace(["", "none"], np.nan)

print("CANTIDAD DE VALORES NULOS")
print(df_clientes.isnull().sum())

df_clientes["ciudad"] = df_clientes["ciudad"].fillna("No especificada")
df_clientes = df_clientes.dropna(subset=["correo"])

print("\nDATAFRAME LIMPIO")
print(df_clientes)
```

# Ejercicio 19: Análisis de préstamos

**Enunciado:** estructurar un DataFrame de solicitudes de crédito para calcular el monto promedio, contar solicitudes por estado, filtrar registros aprobados o con montos altos, y promediar los montos agrupados por estado.

**Lógica:** se crea un DataFrame con pandas. Se usa mean() para el promedio global, value_counts() para las frecuencias de los estados, segmentación lógica para los filtros condicionales, y groupby() junto con mean() para los promedios indexados por estado.

```python
import pandas as pd

datos_prestamos = {
    "cliente": ["Carlos Mendieta", "Ana Cruz", "Luis Valle", "Marta Gómez", "Jorge Paz",
               "Elena Solís", "Diego Reyes", "Laura Núñez", "Sofia Meza", "Ramiro Lobo"],
    "monto_solicitado": [45000, 60000, 30000, 75000, 20000, 90000, 50000, 15000, 110000, 40000],
    "plazo_meses": [36, 48, 24, 60, 12, 60, 36, 12, 72, 24],
    "ingreso_mensual": [25000, 35000, 18000, 40000, 15000, 55000, 28000, 12000, 65000, 22000],
    "estado": ["Aprobado", "Aprobado", "Rechazado", "Pendiente", "Aprobado", 
               "Aprobado", "Rechazado", "Aprobado", "Pendiente", "Rechazado"]
}
df_prestamos = pd.DataFrame(datos_prestamos)

monto_promedio = df_prestamos["monto_solicitado"].mean()
print(f"Monto promedio solicitado: {monto_promedio}")

print("\nSolicitudes por estado:")
print(df_prestamos["estado"].value_counts())

print("\nPréstamos aprobados:")
print(df_prestamos[df_prestamos["estado"] == "Aprobado"])

print("\nClientes con monto solicitado mayor a 50,000:")
print(df_prestamos[df_prestamos["monto_solicitado"] > 50000])

print("\nPromedio de monto solicitado por estado:")
print(df_prestamos.groupby("estado")["monto_solicitado"].mean().reset_index())
```

# Ejercicio 20: Mini análisis de datos tipo empresa 

**Enunciado:** estructurar un DataFrame corporativo para explorar estadísticas salariales, calcular promedios por departamento, identificar el sueldo máximo, filtrar por experiencia, ordenar registros y categorizar el nivel profesional.

**Lógica:** se inicializa el DataFrame con pandas. Se usan head() y describe() para la exploración y estadísticas. Los cálculos y filtros se realizan con groupby(), mean(), idxmax() y segmentación lógica. La columna condicional se genera mediante cut() o loc[] basado en rangos de años.

```python
import pandas as pd

datos_empleados = {
    "nombre": ["Carlos Mendieta", "Ana Cruz", "Luis Valle", "Marta Gómez", "Jorge Paz",
               "Elena Solís", "Diego Reyes", "Laura Núñez", "Sofia Meza", "Ramiro Lobo"],
    "departamento": ["Sistemas", "Sistemas", "RRHH", "Contabilidad", "Sistemas", 
                    "RRHH", "Contabilidad", "Sistemas", "Contabilidad", "RRHH"],
    "cargo": ["Desarrollador", "DBA", "Analista", "Contador", "Soporte", 
              "Gerente", "Asistente", "Desarrollador", "Auditor", "Reclutador"],
    "salario": [45000, 55000, 30000, 42000, 25000, 75000, 22000, 48000, 50000, 28000],
    "experiencia_anios": [4, 6, 2, 8, 1, 12, 1, 5, 7, 3],
    "ciudad": ["Tegucigalpa", "San Pedro Sula", "La Ceiba", "Tegucigalpa", "Comayagua",
               "San Pedro Sula", "Choluteca", "Tegucigalpa", "Danlí", "Santa Rosa"]
}
df_empleados = pd.DataFrame(datos_empleados)

print("PRIMEROS REGISTROS")
print(df_empleados.head(3))

print("\nESTADÍSTICAS GENERALES DE SALARIOS")
print(df_empleados["salario"].describe())

print("\nSALARIO PROMEDIO POR DEPARTAMENTO")
print(df_empleados.groupby("departamento")["salario"].mean())

print("\nEMPLEADO CON EL SALARIO MÁS ALTO")
empleado_top = df_empleados.loc[df_empleados["salario"].idxmax()]
print(empleado_top)

print("\nEMPLEADOS CON MÁS DE 5 AÑOS DE EXPERIENCIA")
print(df_empleados[df_empleados["experiencia_anios"] > 5])

print("\nEMPLEADOS ORDENADOS POR SALARIO")
print(df_empleados.sort_values(by="salario", ascending=False))

df_empleados.loc[df_empleados["experiencia_anios"] <= 2, "NivelExperiencia"] = "Junior"
df_empleados.loc[(df_empleados["experiencia_anios"] > 2) & (df_empleados["experiencia_anios"] <= 5), "NivelExperiencia"] = "Intermedio"
df_empleados.loc[df_empleados["experiencia_anios"] > 5, "NivelExperiencia"] = "Senior"

print("\nDATAFRAME FINAL CON NIVEL DE EXPERIENCIA")
print(df_empleados)
```
